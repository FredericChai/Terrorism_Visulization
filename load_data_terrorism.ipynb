```python
# a word cloud made by Global terrorism
import numpy as np 
import pandas as pd 
import plotly.plotly as py 
import jieba
import plotly.graph_objs as go
from wordcloud import WordCloud
import matplotlib.pyplot as plt

```

```python
dataset = pd.read_csv('/Users/chaizhizhi/Desktop/globalterrorismdb_processed.csv',encoding='ISO-8859-1')
dt = dataset
txt = dt["country_txt"]
word_lst = []
word_dict = {}
for i in range(len(txt)):
    word_lst.append(txt[i])
```

```python
# statistic the word count by the country and write it into a document

for item_1 in word_lst:
        if item_1 not in word_dict:
            word_dict[item_1] = 1
        else:
            word_dict[item_1] += 1
#sort the words according to the frequency
word_dict_sorted = dict(sorted(word_dict.items(), key = lambda item:item[1], reverse=True))

#    Export to a text file
#     for key in word_dict_sorted:
#            wf2.write(key+' '+str(word_dict_sorted[key])+'\n') 
#     wf2.close()
 
```

```python
plt.subplots(figsize=(18,15))
wc = WordCloud(
    font_path = '/Users/chaizhizhi/Desktop/28 Days Later.TTF',
    width = 1980, height = 1280, margin= 0,  
    max_words = 1000, 
    min_font_size = 20,
    max_font_size = 200,
    background_color="White").generate_from_frequencies(word_dict_sorted) 

plt.imshow(wc)
plt.axis("off")
plt.show()
wc.to_file("country_2.jpg")


# wc = WordCloud(
#         #
#         font_path="simkai.ttf",
#         #font_path=path.join(d,'simsun.ttc'),
#         #
#         background_color='white',
#         #
#         mask=color_mask,
#         #
#         max_words=2000,
#         #
#         max_font_size=60
#     )
```
